**Retrieving Stock Data With Pandas**

This is a quick showcase of how to retrieve stock data using the *yfinance* extension. Here we retrieve stock data for AAPL, NVDA, TSLA and SXR8.

```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance
```

*Specify stocks*

```python
stocklist = ["AAPL", "NVDA", "SXR8.DE", "TSLA"]
```

```python
data = yfinance.download(stocklist, start="2014-09-20", end="2024-09-20")
```

```python
percentage_change_NVDA = []
for i in range(2581):
    change = (close_price["NVDA"].iloc[i+1] - close_price["NVDA"].iloc[i]) / close_price["NVDA"].iloc[i]
    percentage_change_NVDA.append(change)
data_changed.insert(1, "NVDA change", percentage_change_NVDA)

percentage_change_TSLA = []
for i in range(2581):
    change = (close_price["TSLA"].iloc[i+1] - close_price["TSLA"].iloc[i]) / close_price["TSLA"].iloc[i]
    percentage_change_TSLA.append(change)
data_changed.insert(2, "TSLA change", percentage_change_TSLA)

percentage_change_SXR8 = []
for i in range(2581):
    change = (close_price["SXR8.DE"].iloc[i+1] - close_price["SXR8.DE"].iloc[i]) / close_price["SXR8.DE"].iloc[i]
    percentage_change_SXR8.append(change)
data_changed.insert(3, "SXR8.DE change", percentage_change_SXR8)

data_changed
```

**10 year stock returns for AAPL, NVDA, TSLA and SXR8.DE (in %)**

```python
print(int(((1 + data_changed["AAPL change"]).prod() - 1) * 100), "%")
print(int(((1 + data_changed["NVDA change"]).prod() - 1) * 100), "%")
print(int(((1 + data_changed["TSLA change"]).prod() - 1) * 100), "%")
print(int(((1 + data_changed["SXR8.DE change"]).prod() - 1) * 100), "%")
```
