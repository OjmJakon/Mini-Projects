**Retrieving Stock Data With Pandas**

This is a quick showcase of how to retrieve stocl data using the *yfinance* extension. Here we retrieve stock data for AAPL, NVDA, TSLA and SXR8.

```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance
```

*Specify stocks*

```python
stocklist = ["AAPL", "NVDA", "SXR8.DE", "TSLA"]
```

```python
data = yfinance.download(stocklist, start="2014-09-20", end="2024-09-20")
```
